```python
import pandas as pd
import yfinance as yf

companies = ['AAPL', 'AMZN', 'GOOG', 'META', 'NFLX', 'MSFT']


date_ranges = {
    'Q3': ('2024-07-30', '2024-09-30'),
    'Q4': ('2024-10-01', '2024-12-31')
}

all_combined_data = pd.DataFrame()

def download_and_clean_data(tickers, start_date, end_date, quarter):
    combined_data = pd.DataFrame()
    for company in tickers:
        print(f"Downloading data for {company} for {quarter} 2024...")

        # Download the stock data for the given date range
        data = yf.download(company, start=start_date, end=end_date)

        # Flatten MultiIndex columns if necessary and ensure 'Date' is part of the DataFrame
        if isinstance(data.columns, pd.MultiIndex):
            data.columns = ['_'.join(tup).strip() for tup in data.columns]

        # Ensure 'Date' is part of the DataFrame
        if 'Date' not in data.columns and data.index.name == 'Date':
            data.reset_index(inplace=True)

        if 'Date' not in data.columns:
            print(f"Error: 'Date' column missing in {company}'s data after reset. Skipping...")
            continue
        else:
            data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

        # Rename columns appropriately
        data.columns = ['Date' if col == 'Date' else col.replace(f"_{company}", "") for col in data.columns]
        data.columns = [f"{col}_{company}" if col != 'Date' else col for col in data.columns]

        # Validate required columns
        required_columns = ['Date', f'Open_{company}', f'High_{company}', f'Low_{company}', f'Close_{company}', f'Volume_{company}']
        if not all(col in data.columns for col in required_columns):
            print(f"Warning: Missing columns in {company}'s data. Found columns: {list(data.columns)}. Skipping...")
            continue

        # Filter data for the given date range
        data = data[data['Date'].between(start_date, end_date)]

        # Print first few rows for verification
        print(f"Data for {company} after processing: {data.head()}")

        # Merge data into a combined DataFrame
        if combined_data.empty:
            combined_data = data.copy()
        else:
            combined_data = pd.merge(combined_data, data, on='Date', how='outer')

        print(f"Data for {company} ({quarter} 2024) downloaded and cleaned successfully!")

    # Save combined data for the quarter to CSV
    combined_data.to_csv(f'combined_{quarter}_2024_cleaned_stock_data.csv', index=False)
    print(f'Combined data for {quarter} 2024 saved successfully!')
    return combined_data

# Iterate over date ranges for Q3 and Q4
for quarter, (start_date, end_date) in date_ranges.items():
    combined_quarter_data = download_and_clean_data(companies, start_date, end_date, quarter)
    all_combined_data = pd.concat([all_combined_data, combined_quarter_data])

# Save the combined data to a final CSV file
all_combined_data.to_csv('combined_Q3_Q4_2024_cleaned_stock_data.csv', index=False)
print("Final CSV file for combined Q3 and Q4 data has been successfully saved!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

data = pd.read_csv('combined_Q3_Q4_2024_cleaned_stock_data.csv')

data['Date'] = pd.to_datetime(data['Date'])


data.set_index('Date', inplace=True)
```

```python
def analyze_stock(stock_symbol):

    data[f'Close_{stock_symbol}'] = data[f'Close_{stock_symbol}'].interpolate(method='linear').ffill()


    if data[f'Close_{stock_symbol}'].isnull().sum() == 0:

        resampled_data = data[f'Close_{stock_symbol}'].resample('D').mean()
        period = 52
        decomposition = seasonal_decompose(resampled_data.dropna(), model='additive', period=52)
        decomposition.plot()
        plt.title(f'Seasonal Decomposition of {stock_symbol} Stock Prices')
        plt.show()


        data[f'{stock_symbol}_Rolling'] = data[f'Close_{stock_symbol}'].rolling(window=30).mean()

        plt.figure(figsize=(12, 6))
        plt.plot(data[f'Close_{stock_symbol}'], label='Original')
        plt.plot(data[f'{stock_symbol}_Rolling'], label='30 Days Rolling Mean')
        plt.title(f'{stock_symbol} Stock Price with 30 Days Rolling Mean')
        plt.legend()
        plt.show()
    else:
        print(f"There are still missing values in {stock_symbol}'s dataset.")
```

```python

stocks = ['AMZN', 'GOOG', 'META', 'NFLX', 'MSFT']

for stock in stocks:
    analyze_stock(stock)
```

```python
def fit_arima_model(stock_data, stock_symbol, order=(5, 1, 0)):

    train_size = int(len(stock_data) * 0.8)
    train, test = stock_data[:train_size], stock_data[train_size:]


    model = ARIMA(train, order=order)
    model_fit = model.fit()


    predictions = model_fit.forecast(steps=len(test))


    mae = mean_absolute_error(test, predictions)
    print(f'{stock_symbol} - Mean Absolute Error: {mae}')


    plt.figure(figsize=(12, 6))
    plt.plot(train, label='Training Data')
    plt.plot(test, label='Actual Stock Prices')
    plt.plot(test.index, predictions, label='Predicted Stock Prices')
    plt.title(f'{stock_symbol} Stock Price Prediction')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()

    return model_fit, predictions
```

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Function to find the best ARIMA model
def find_best_arima(data, p_values, d_values, q_values):
    best_mae = float('inf')
    best_order = None
    best_model_fit = None

    for p in p_values:
        for d in d_values:
            for q in q_values:
                try:
                    model = ARIMA(data['Close_AAPL'], order=(p, d, q))
                    model_fit = model.fit()
                    predictions = model_fit.forecast(steps=len(data['Close_AAPL']) // 2)
                    mae = mean_absolute_error(data['Close_AAPL'][len(data) // 2:], predictions)
                    if mae < best_mae:
                        best_mae = mae
                        best_order = (p, d, q)
                        best_model_fit = model_fit
                except:
                    continue

    return best_order, best_model_fit

# Define range of parameters
p_values = range(0, 5)
d_values = range(0, 2)
q_values = range(0, 5)

# Find the best ARIMA model
best_order, best_model_fit = find_best_arima(data, p_values, d_values, q_values)
print(f'Best ARIMA order: {best_order}')

# Make predictions using the best model
predictions = best_model_fit.forecast(steps=len(data['Close_AAPL']) // 2)

# Evaluate predictions
mae = mean_absolute_error(data['Close_AAPL'][len(data) // 2:], predictions)
print(f'Mean Absolute Error: {mae}')
mse = mean_squared_error(data['Close_AAPL'][len(data) // 2:], predictions)
print(f'Mean Squared Error: {mse}')

# Plot the predictions against actual values
plt.figure(figsize=(12, 6))
plt.plot(data['Close_AAPL'], label='Actual Stock Prices')
plt.plot(data.index[len(data)//2:], predictions, label='Predicted Stock Prices', color='red')
plt.title('Actual vs Predicted Stock Prices')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()

# Diagnostic plots
best_model_fit.plot_diagnostics(figsize=(12, 8))
plt.show()
```

```python
def train_evaluate_arima(data, stock_symbol, order, seasonal_order):
    stock_data = data[['Close_' + stock_symbol]].dropna()
    stock_data.columns = ['Close']

    # Split the data into training and testing sets
    train_data = stock_data[:int(0.7 * len(stock_data))]
    test_data = stock_data[int(0.7 * len(stock_data)):]

    # Fit the SARIMAX model
    model = SARIMAX(train_data['Close'], order=order, seasonal_order=seasonal_order)
    model_fit = model.fit(disp=False)

    # Make predictions
    predictions = model_fit.predict(start=len(train_data), end=len(stock_data) - 1, dynamic=False)

    # Evaluate the model
    mae = mean_absolute_error(test_data['Close'], predictions)
    print(f'{stock_symbol} - ARIMA Model MAE: {mae}')

    # Plot the actual vs predicted values
    plt.figure(figsize=(12, 6))
    plt.plot(test_data.index, test_data['Close'], label='Actual Values')
    plt.plot(test_data.index, predictions, label='Predicted Values', color='red')
    plt.title(f'Actual vs Predicted Values for {stock_symbol}')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()

    return model_fit, predictions
```

```python
import pandas as pd
import numpy as np
import pmdarima as pm
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt

# Load the combined stock data
data = pd.read_csv('/content/combined_Q3_2024_cleaned_stock_data.csv')

# Ensure the date column is in datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Set the date column as the index
data.set_index('Date', inplace=True)
```

```python
# Extract META stock data
meta_data = data[['Close_META']].dropna()
meta_data.columns = ['Close']

# Split the data into training and testing sets
train_meta_data = meta_data[:int(0.7 * len(meta_data))]
test_meta_data = meta_data[int(0.7 * len(meta_data)):]

# Use Auto-ARIMA to find the best ARIMA parameters
auto_arima_model = pm.auto_arima(train_meta_data['Close'], seasonal=True, m=12, trace=True, error_action='ignore', suppress_warnings=True)

# Display the best parameters
print(f'Best ARIMA parameters for META: {auto_arima_model.order}, {auto_arima_model.seasonal_order}')

# Fit the best ARIMA model
auto_arima_model.fit(train_meta_data['Close'])

# Make predictions
meta_predictions = auto_arima_model.predict(n_periods=len(test_meta_data))

# Evaluate the model
meta_mae = mean_absolute_error(test_meta_data['Close'], meta_predictions)
print(f'META - Auto-ARIMA Model MAE: {meta_mae}')

# Plot the actual vs predicted values
plt.figure(figsize=(12, 6))
plt.plot(test_meta_data.index, test_meta_data['Close'], label='Actual Values')
plt.plot(test_meta_data.index, meta_predictions, label='Predicted Values', color='red')
plt.title('Actual vs Predicted Values for META using Auto-ARIMA')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
import numpy as np
from statsmodels.tsa.stattools import adfuller

# Extract META stock data
meta_data = data[['Close_META']].dropna()
meta_data.columns = ['Close']

# Check for stationarity using the Augmented Dickey-Fuller test
adf_result = adfuller(meta_data['Close'])
print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')

# Apply log transformation if the series is not stationary
if adf_result[1] > 0.05:
    meta_data['Log_Close'] = np.log(meta_data['Close'])
    log_adf_result = adfuller(meta_data['Log_Close'])
    print(f'Log ADF Statistic: {log_adf_result[0]}')
    print(f'Log p-value: {log_adf_result[1]}')
    transformed_data = meta_data['Log_Close']
else:
    transformed_data = meta_data['Close']
```

```python
# Apply first-order differencing
meta_data['Log_Close_Diff'] = meta_data['Log_Close'].diff().dropna()

# Check for stationarity again
diff_adf_result = adfuller(meta_data['Log_Close_Diff'].dropna())
print(f'Differenced ADF Statistic: {diff_adf_result[0]}')
print(f'Differenced p-value: {diff_adf_result[1]}')

# If differencing makes the series stationary, use the differenced series for modeling
if diff_adf_result[1] <= 0.05:
    transformed_data = meta_data['Log_Close_Diff'].dropna()
else:
    transformed_data = meta_data['Log_Close']
```

```python
# Split the differenced data into training and testing sets
train_meta_data = transformed_data[:int(0.7 * len(transformed_data))]
test_meta_data = transformed_data[int(0.7 * len(transformed_data)):]

# Use Auto-ARIMA to find the best ARIMA parameters
auto_arima_model = pm.auto_arima(train_meta_data, seasonal=True, m=12, trace=True, error_action='ignore', suppress_warnings=True)

# Display the best parameters
print(f'Best ARIMA parameters for META: {auto_arima_model.order}, {auto_arima_model.seasonal_order}')

# Fit the best ARIMA model
auto_arima_model.fit(train_meta_data)

# Make predictions
meta_predictions = auto_arima_model.predict(n_periods=len(test_meta_data))

# Evaluate the model
meta_mae = mean_absolute_error(test_meta_data, meta_predictions)
print(f'META - Auto-ARIMA Model MAE: {meta_mae}')

# Plot the actual vs predicted values
plt.figure(figsize=(12, 6))
plt.plot(test_meta_data.index, test_meta_data, label='Actual Values')
plt.plot(test_meta_data.index, meta_predictions, label='Predicted Values', color='red')
plt.title('Actual vs Predicted Values for META using Auto-ARIMA on Differenced Series')
plt.xlabel('Date')
plt.ylabel('Differenced Log Stock Price')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt

# Load the combined data
data = pd.read_csv('/content/combined_Q3_Q4_2024_cleaned_stock_data.csv')

# Define features (X) and target (y)
companies = ['GOOG', 'META', 'NFLX', 'MSFT', 'AMZN']
X = data.drop(columns=[f'Close_{company}' for company in companies] + ['Date'])
y = data[[f'Close_{company}' for company in companies]]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define models and pipelines
pipelines = {
    'Linear Regression': Pipeline([
        ('scaler', StandardScaler()),
        ('regressor', LinearRegression())
    ]),
    'Random Forest': Pipeline([
        ('scaler', StandardScaler()),
        ('regressor', RandomForestRegressor())
    ]),
    'XGBoost': Pipeline([
        ('scaler', StandardScaler()),
        ('regressor', XGBRegressor(n_estimators=50, max_depth=3))  # Using best parameters found
    ])
}

# Define hyperparameters
params = {
    'Linear Regression': {},
    'Random Forest': {
        'regressor__n_estimators': [50, 100, 200],
        'regressor__max_depth': [3, 5, 10, None]
    },
    'XGBoost': {
        'regressor__n_estimators': [50, 100, 200],
        'regressor__max_depth': [3, 5, 10]
    }
}

# Perform Grid Search
best_models = {}
for model_name in pipelines:
    grid_search = GridSearchCV(pipelines[model_name], params[model_name], cv=5, scoring='neg_mean_absolute_error')
    grid_search.fit(X_train, y_train)
    best_models[model_name] = grid_search.best_estimator_
    print(f'Best parameters for {model_name}: {grid_search.best_params_}')

# Evaluate Models
for model_name in best_models:
    model = best_models[model_name]
    predictions = model.predict(X_test)
    mae = mean_absolute_error(y_test, predictions)
    mse = mean_squared_error(y_test, predictions)
    print(f'{model_name} - Mean Absolute Error: {mae}, Mean Squared Error: {mse}')

    # Plot the actual vs predicted values
    plt.figure(figsize=(12, 6))
    plt.plot(y_test.values, label='Actual Values')
    plt.plot(predictions, label='Predicted Values', color='red')
    plt.title(f'Actual vs Predicted Values for {model_name}')
    plt.xlabel('Sample Index')
    plt.ylabel('Target Value')
    plt.legend()
    plt.show()
```
